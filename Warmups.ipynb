```python
## Warmup 1, Feb 15

import array
import numpy as np

## Part 2
def perceptron_classify(w, b, x):
    if w.dot(x) < b:
        return (0)
    else:
        return (1)

w = np.array([5, 4])
x = np.array([1, 1])
b = 10

print (perceptron_classify(w, b, x))

def perceptron_classify_revised(w, b, x):
    return 1*(w@x - b > 0)
## 1* boolean T & F turns into 1 & 0 
```

```python
## Warmup 2, Feb 20

from matplotlib import pyplot as plt 
import numpy as np

def f(z):
   return (-np.log(z))

def g(z):
   return (-np.log(1-z))

z = np.linspace(-10, 10, 100)

plt.plot(z, f(z), color='red')
plt.plot(z, g(z), color='blue')

plt.show()
```
